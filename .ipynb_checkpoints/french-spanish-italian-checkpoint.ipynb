```python
import pandas as pd

csv_file = './data/only_english2.csv'
df = pd.read_csv(csv_file, encoding='utf-8')

French = df[df['native'].str.contains('French')]
print(len(French))
Spanish = df[df['native'].str.contains('Spanish')]
print(len(Spanish))
Italian = df[df['native'].str.contains('Italian')]
print(len(Italian))

languages = {'French': French, 'Spanish': Spanish, 'Italian': Italian}
```

```python
import pandas as pd
from happytransformer import HappyTextToText, TTSettings
from tqdm import tqdm

happy_tt = HappyTextToText("T5", "prithivida/grammar_error_correcter_v1")

def correction(language, df):
    text_list = df["text"].tolist()
    setting = TTSettings(do_sample=True, top_k=10, temperature=0.5, min_length=1, max_length=100)

    original_texts = []
    corrected_texts = []

    for text in tqdm(text_list, desc="Processing texts"):
        original_texts.append(text)
        corrected_text = happy_tt.generate_text(f"gec: {text}", args=setting).text
        corrected_texts.append(corrected_text)

    output_df = pd.DataFrame({'original': original_texts, 'modified': corrected_texts})

    # Use double backslashes or a raw string
    output_csv_file = './data/' + language + '.csv'
    output_df.to_csv(output_csv_file, index=False, encoding='utf-8')

    print(f"CSV file with corrected data saved to {output_csv_file}")
```

```python
for key, val in languages.items():
    correction(key, val)
```

```python
language_files = {'French': './data/French.csv', 'Spanish': './data/Spanish.csv', 'Italian': './data/Italian.csv'}
```

```python
import pandas as pd
from tqdm import tqdm
import errant
import spacy
import seaborn as sns
import matplotlib.pyplot as plt

nlp = spacy.load('en_core_web_sm')
annotator = errant.load('en', nlp)

def types_and_frequencies(language, file):
    df = pd.read_csv(file)
    error_types = []
    temp_data = df.head(5)
    for i in tqdm(range(len(df))):
        temp_data = df.iloc[i]
        orig = annotator.parse(temp_data['original'])
        cor = annotator.parse(temp_data['modified'])
        edits = annotator.annotate(orig, cor)
        for e in edits:
            error_types.append(e.type)
            #print(e.o_start, e.o_end, e.o_str, e.c_start, e.c_end, e.c_str, e.type)

    error_df = pd.DataFrame(columns = ['type'], data=error_types)
    #error_df.head()

    #error_df.value_counts(normalize=True)

    freq = pd.DataFrame(error_df.value_counts(), columns=['count'])
    freq = freq.reset_index()
    #freq.head()

    plt.figure(figsize=(20,8))
    pl = sns.barplot(x='type', y='count', data=freq)
    plt.xticks(rotation=45)

    pl.axes.set_yscale('log')
    plt.title(language)
    plt.savefig('./plots/' + language + '.png')
    plt.show()
```

```python
for language, file in language_files.items():
    types_and_frequencies(language, file)
```
